```python
import numpy as np
import matplotlib.pyplot as plt
```

```python




nsteps = 100000

a = 0.6
b = 1.2
c = 1
d = 1

possible_ds = [0.5,1,1.5,2] # all possible values for 'd'
dt = 0.001

x = np.zeros(nsteps)
y = np.zeros(nsteps)
for d in possible_ds:
    x[0] = 1
    y[0] = 1

    for t in range(1,nsteps):
        x[t] = x[t-1] + dt*(a*x[t-1]-b*x[t-1]*y[t-1])
        y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1])
    
    
#first plot
    fig1 = plt.figure();
    plt.plot(x,label='Bunnies');
    plt.plot(y,label='Foxes');
    plt.legend(loc='upper right');

#second plot
    fig2 = plt.figure(figsize=(5,5));
    
    plt.plot(x,y, label=d);
    plt.xlabel('Bunnies');
    plt.ylabel('Foxes');
```
