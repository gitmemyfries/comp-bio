```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib notebook

n_animals = 100
n_steps = 1000
diffusion_speed = 0.01

x = np.ones(n_animals)*.5
y = np.ones(n_animals)*.5


fig = plt.figure(figsize=(3,3));

for i in range(n_steps):
    x+=np.random.randn(n_animals)*diffusion_speed
    y+=np.random.randn(n_animals)*diffusion_speed
    for t in range(n_animals):
        if x[t] < 0:
            x[t] = 0
        if x[t] > 1:
            x[t] = 1
        if  y[t] < 0:
            y[t] = 0
        if y[t] > 1:
            y[t] = 1
    plt.cla()
    plt.xlim((0,1));
    plt.ylim((0,1));
    plt.plot(x,y,'.k', markersize=2);
    fig.canvas.draw()
    time.sleep(0.01)
  

```
